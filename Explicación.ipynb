#Contexto

Se tiene un conjunto de datos que contienen información sobre diferentes casas como el codigo postal, las dimensiones de la casa, la ubicación en coordenadas y el año de construcción. En base a estos datos se crea un modelo de predicción donde el valor a predecir es el precio que podria llegar a tener una determinada casa. Todo este proceso teniendo en cuenta un análisis y limpieza de los datos, la prueba de distintos modelos y el chequeo de residuos generado.

#Información del conjunto de datos:

• id: un número entre 0 y 100000 que es un identificador único para cada casa. No uses este dato en tus algoritmos.

•zipcode: cada casa tiene un código postal correspondiente al área en la que se encuentra. El código postal puede indicar el año de construcción u otras propiedades. Hay cuatro códigos postales en uso: 3520, 3525, 3525 y 3800.

•lot-len: la longitud en metros del terreno en el que está construida la casa. Cada casa está construida en un terreno cuadrado. Puede variar entre 5.0 y 100.0 metros.

•lot-width: el ancho en metros del terreno en el que está construida la casa. Puede variar entre 5.0 y 100.0 metros.

•lot-area: el área total del terreno en el que está construida la casa. Probablemente se pueda calcular a partir de lot-len y lot-width.

•house-area: el área habitable de la casa en metros cuadrados. 30.0 metros cuadrados es una casa pequeña, 200.0 metros cuadrados sería una mansión.

•garden-size: el tamaño del jardín en metros cuadrados. Muchas personas desean tener un jardín grande.

•balcony: el número de balcones que tiene la casa. Los valores comunes son 0, 1 o 3 balcones.

•x-coor: la coordenada x que describe la ubicación de la casa. Es un valor entero entre 2000 y 3000.

•y-coor: la coordenada y que describe la ubicación de la casa. Es un valor entero entre 5000 y 6000.

•buildyear: el año en que se construyó la casa. Algunas de las casas más antiguas son del año 1100, pero la mayoría de las casas fueron construidas en los siglos XX y XXI
